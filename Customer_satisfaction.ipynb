```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
```

```python
# Reading dataset
train_data = pd.read_csv('train.csv')
test_data = pd.read_csv('test.csv')
```

- Deleting all columns with no valuable information ('Unnamed: 0' and 'id')

```python
train_data = train_data.drop(['Unnamed: 0','id'], axis = 1)
test_data = test_data.drop(['Unnamed: 0','id'], axis = 1)
```

- converting all categorial data into numeric

```python
le = LabelEncoder()
train_data['Gender'] = le.fit_transform(train_data['Gender'])
train_data['Customer Type'] = le.fit_transform(train_data['Customer Type'])
train_data['Type of Travel'] = le.fit_transform(train_data['Type of Travel'])
train_data['Class'] = le.fit_transform(train_data['Class'])
train_data['satisfaction'] = le.fit_transform(train_data['satisfaction'])
```

```python
train_data.head(5)
```

```python
le = LabelEncoder()
test_data['Gender'] = le.fit_transform(test_data['Gender'])
test_data['Customer Type'] = le.fit_transform(test_data['Customer Type'])
test_data['Type of Travel'] = le.fit_transform(test_data['Type of Travel'])
test_data['Class'] = le.fit_transform(test_data['Class'])
test_data['satisfaction'] = le.fit_transform(test_data['satisfaction'])
```

```python
test_data.head(5)
```

Checking for null values

```python
train_data.isnull().sum()
```

```python
train_data.isnull().sum()
```

Filling all the null values with median 

```python
train_data['Arrival Delay in Minutes'] = train_data['Arrival Delay in Minutes'].fillna(train_data['Arrival Delay in Minutes'].median())
test_data['Arrival Delay in Minutes'] = test_data['Arrival Delay in Minutes'].fillna(test_data['Arrival Delay in Minutes'].median())
```

```python
train_data.isnull().sum()
```

```python
test_data.isnull().sum()
```

Determining which features are skewed.

```python
numeris_feature = train_data.select_dtypes(include=[np.number]).columns
skewness = train_data[numeris_feature].skew().sort_values(ascending=False)
features_to_log = skewness[skewness > 1].index.tolist()
print(features_to_log)
```

Apply logarithm to them

```python
for features in features_to_log:
    train_data[features] = np.log1p(train_data[features])
    test_data[features] = np.log1p(test_data[features])
```

split the dataset into features and target.

```python
x_train = train_data.drop(columns= ['satisfaction'])
y_train = train_data['satisfaction'].to_numpy()
x_test = test_data.drop(columns= ['satisfaction'])
y_test = test_data['satisfaction'].to_numpy()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score, GridSearchCV
from sklearn.metrics import accuracy_score
```

```python
from sklearn.metrics import f1_score, roc_auc_score


def evaluate_model(model, x_test, y_test):
    y_pred = model.predict(x_test)
    y_probability = model.predict_proba(x_test)[:,1]
    accuracy = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred)
    
    print(f'Accuracy: {accuracy} \nF1-Score:{f1} \nROC-AUC:{roc_auc}')
    
```

```python
# Vanilla Decision Tree
dt_model = DecisionTreeClassifier()
dt_model.fit(x_train,y_train)
print('Vanilla Decision Tree\n')
print('Metrices on train data:')
print('---'*8)
evaluate_model(dt_model,x_train,y_train)
print('\nMetrices on test:')
print('--'*9)
evaluate_model(dt_model,x_test,y_test)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve

def plot_roc_curve(model, x_test, y_test, model_name):
    y_proba = model.predict_proba(x_test)[:, 1]
    
    fpr, tpr, _ = roc_curve(y_test, y_proba)
    auc_score = roc_auc_score(y_test, y_proba)
    
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, color='salmon', label=f'ROC Curve (AUC = {auc_score:.4f})')
    plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Diagonal line for random guessing
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'ROC Curve - {model_name}')
    plt.legend(loc='lower right')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()

plot_roc_curve(dt_model, x_test, y_test, 'Vanilla Decision Tree')
plot_roc_curve(best_dt_model, x_test, y_test, 'Decision Tree with CV')
```

```python
# Vanilla Decision Tree
y_proba_dt = dt_model.predict_proba(x_test)[:, 1]
fpr_dt, tpr_dt, _ = roc_curve(y_test, y_proba_dt)
auc_score_dt = roc_auc_score(y_test, y_proba_dt)

# Decision Tree with CV
y_proba_dtcv = best_dt_model.predict_proba(x_test)[:, 1]
fpr_dtcv, tpr_dtcv, _ = roc_curve(y_test, y_proba_dtcv)
auc_score_dtcv = roc_auc_score(y_test, y_proba_dtcv)

# both ROC curves
plt.figure(figsize=(8, 6))
plt.plot(fpr_dt, tpr_dt, color='salmon', label=f'Vanilla Decision Tree (AUC = {auc_score_dt:.4f})', linewidth=2)
plt.plot(fpr_dtcv, tpr_dtcv, color='cyan', label=f'Decision Tree with CV (AUC = {auc_score_dtcv:.4f})', linewidth=2)

# random guessing
plt.plot([0, 1], [0, 1], color='gray', linestyle='--', linewidth=1.5)

plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate', fontsize=12)
plt.title('ROC Curve Comparison', fontsize=16)
plt.legend(loc='lower right', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
